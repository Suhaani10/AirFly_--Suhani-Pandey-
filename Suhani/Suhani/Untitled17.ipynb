```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
print("Shape of dataset:", df.shape)
print("\nColumn Names:\n", df.columns.to_list())
print("\nInfo:\n")
df.info()
print("\nMissing Values:\n", df.isnull().sum())
print("\nStatistical Summary (Numerical Columns):\n", df.describe())
```

```python
df.drop_duplicates(inplace=True)


df.drop(columns=['First Name', 'Last Name', 'Pilot Name'], inplace=True, errors='ignore')


num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[num_cols] = num_imputer.fit_transform(df[num_cols])
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])


df['Departure Date'] = pd.to_datetime(df['Departure Date'], errors='coerce')
```

```python
# --- 1. Check for missing values ---
print("Missing values before cleaning:", df['Departure Date'].isna().sum())

# --- 2. Drop rows where Departure Date is missing or blank ---
df = df[df['Departure Date'].notna()]              # removes NaN
df = df[df['Departure Date'].str.strip() != '']    # removes empty strings

# --- 3. Convert Departure Date to datetime (if not already) ---
df['Departure Date'] = pd.to_datetime(df['Departure Date'], errors='coerce')

# --- 4. Drop rows where conversion failed (invalid dates) ---
df = df.dropna(subset=['Departure Date'])

# --- 5. Remove duplicate dates (if same date entries are repeated unnecessarily) ---
df = df.drop_duplicates(subset=['Departure Date'])

# --- 6. Verify the cleanup ---
print("Missing values after cleaning:", df['Departure Date'].isna().sum())
print("Total unique dates:", df['Departure Date'].nunique())
```
